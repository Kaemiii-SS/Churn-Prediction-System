```python
df["TotalCharges"] = pd.to_numeric(df["TotalCharges"], errors="coerce")
```

```python
df["TotalCharges"].isna().sum()
```

TotalCharges contained 11 missing values (≈0.13% of the dataset). Since the proportion was very small and the feature is numerical, missing values were imputed using the median to minimize the influence of outliers.

```python
df.fillna({"TotalCharges": df["TotalCharges"].median()}, inplace=True)
```

```python
df["TotalCharges"].isna().sum()
```

yeah so now the plots are readable

### Observations
1. Customers with lower tenure are more likely to churn, while customers who have stayed longer tend to remain with the company. This suggests customer loyalty increases with time.
2. Customers with higher monthly charges appear more likely to churn. This may indicate price sensitivity among customers facing higher recurring costs.
3. TotalCharges ≈ MonthlyCharges × tenure. So Customers with lower TotalCharges tend to churn more frequently, likely because many churning customers are relatively new customers who have not accumulated substantial charges yet.
